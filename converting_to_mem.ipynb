```python
from PIL import Image

img = Image.open("sprite_00.png")
img = img.convert("RGB")

# Force image to 64x64
img = img.resize((64, 64), Image.NEAREST)

width, height = img.size

print(width, height)   # should print 64 64

with open("food_00.mem", "w") as f:
    for y in range(height):
        for x in range(width):
            r, g, b = img.getpixel((x, y))

            # Convert RGB888 -> RGB444
            r4 = r >> 4
            g4 = g >> 4
            b4 = b >> 4

            value = (r4 << 8) | (g4 << 4) | b4

            f.write(f"{value:03X}\n")

print("Done!")
```
